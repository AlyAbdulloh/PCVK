4. Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image
dengan tipe data int16.

```python
black_img = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(black_img)
```

4. 1. Kemudian menambahkan bentuk persegi panjang sesuai koordinat pt1 dan pt2

```python
cv.rectangle(black_img, pt1=(384, 0), pt2=(510, 150), color=(0, 255, 0), thickness=10)
plt.imshow(black_img)
```

4. 2. Selanjutnya ditambah menambahkan bentuk persegi sesuai koordinat pt1 dan pt2
yang tertulis pada kode program.

```python
cv.rectangle(black_img, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness=10)
plt.imshow(black_img)
```

4. 3. Tahap selanjutnya ditambah menambahkan bentuk lingkaran sesuai radius yang
tertulis pada kode program.

```python
cv.circle(black_img, center=(100,100), radius=50, color=(255,0,0), thickness=10)
plt.imshow(black_img)
```

4. 4. Kemudian dilakukan penambahan garis sesuai koordinat pt1 dan pt2 sebagai berikut.

```python
cv.line(black_img, pt1=(0,0), pt2=(512,512), color=(255,255,255), thickness=5)
plt.imshow(black_img)
```

4. 5. Penambahan text dengan font yang telah tertulis dengan ukuran yang sudah
ditentukan.

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Hello', org=(10,500), fontFace=font, fontScale=4, color=(255,255,0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_img)
```

4. 6. Pembuatan black image kembali dilakukan dengan tipe data int32


```python
black_img2 = np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(black_img2)
```

4. 6. Berikut adalah kode program untuk inisialisasi NumPy array dengan tipe data int32

```python
vertices = np.array([[100,300], [200,200], [400,300], [200,400]], dtype=np.int32)
vertices
```

4. 6.1 Array tersebut kemudian di reshape sebagai berikut

```python
pts = vertices.reshape((-1,1,2))
pts
```

4. 7. Penambahan polyline pada black image kedua yang telah dibuat.


```python
cv.polylines(black_img2, [pts], isClosed=True, color=(255,0,0), thickness=5)
plt.imshow(black_img2)
```
